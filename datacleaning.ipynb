```python
#Finding count of null values by columns

sales.isna().sum()
```

```python
#There are issue in the cusomer_name, let's fine the unique customer name

sales['customer_name'].unique()

```

```python
# Removing *,? and extra space from customer name

sales['customer_name']= sales['customer_name'].str.replace('*','').str.replace('?','').str.strip()
sales

# We can also use regex here
```

```python
# unique product
sales['product'].unique()

#Fixsing product name, Removing *,? and extra space

sales['product']= sales['product'].str.replace('*','').str.replace('?','').str.strip()
sales
```

```python
# #unique categories

sales['category'].unique()

# normalising category and remving extra space

sales['category']= sales['category'].str.capitalize().str.strip()

# Catgories are mismatched for Keyboard and Mouse it should be Hardware and Electronics for the rest

sales.loc[sales['product'].str.contains('Keyboard|Mouse', case=False, na=False), 'category'] = 'Hardware'
sales.loc[sales['product'].str.contains('Monitor|Laptop|iPhone', case=False, na=False), 'category'] = 'Electronics'
```

```python
# Quantity has null and negative, making it 0

sales['quantity'].unique() 

sales = sales.fillna({'quantity':0}) # filled null values with 0

sales['quantity']= sales['quantity'].clip(lower=0)  # replace negative with 0
sales
```

```python
sales
```

```python
sales.loc[(sales['quantity'].isna() | (sales['quantity'] == 0)) & 
          (sales['total_amount'] > 0) & (sales['price'] > 0), 
          'quantity'] = sales['total_amount'] / sales['price']
```

```python
sales['total_amount']=sales['total_amount'].round()
```

```python
sales.loc[(sales['price'].isna() | (sales['price'] == 0)) & 
          (sales['total_amount'] > 0) & (sales['quantity'] > 0), 
          'price'] = sales['total_amount'] / sales['quantity']
```
